# **1. Project Overview**

Customer churn is a major challenge in the telecom industry, as it occurs when customers stop using a company’s services or cancel their subscriptions. Retaining existing customers is more cost-effective than acquiring new ones, which makes churn analysis an important task for business growth and sustainability.

This project focuses on analyzing telecom customer data to understand customer behavior, identify the key factors that lead to churn, examine usage and billing patterns, and generate data-driven insights. The analysis includes data preprocessing, statistical and probability analysis, exploratory data analysis, visualization, and correlation study to support effective decision-making.

The ultimate goal of this project is to help the telecom company reduce customer churn, improve customer retention strategies, and enhance long-term business performance through meaningful data analysis.

# 2. **Project objective**

The main objective of this project is to perform a complete data analysis of a telecom customer churn dataset in order to understand the patterns and factors assocated with customer churn. The project aims to explore and preprocess the dataset, apply statistical and probability analysis, and conduct exploratory data analysis to discover meaningful trends in customer behavior.

In addition, the project focuses on creating clear visualizations, identifying high-risk customers who are more likely to churn, and determining the most important features influencing customer retention. These insights are then used to develop practical, data-driven recommendations that can help the telecom company reduce churn and improve long-term customer satisfaction and business performance.

#3.  **Dataset Description**

The dataset used in this project was obtained from Kaggle and contains real-world telecom customer information related to customer subscriptions and churn behavior. The specific file used for analysis is WA_Fn-UseC_-Telco-Customer-Churn.csv, which provides structured data suitable for statistical analysis, visualization, and business insight generation.

This dataset includes important customer attributes such as demographic details, tenure with the company, type of services subscribed, contract type, payment method, monthly charges, total charges, and the churn status indicating whether a customer has left the company or continued using the services.

The availability of these features makes the dataset highly useful for understanding customer behavior, identifying patterns associated with churn, and developing data-driven strategies to improve customer retention in the telecom industry.

# **10. Probability Analysis**

**10.1 churn probability for month_to_month contract**

```python
month_customers = df[(df['Contract_One year'] == 0) & (df['Contract_Two year'] == 0)]
churned_month = month_customers[month_customers['Churn'] == 1]
probability = len(churned_month) / len(month_customers)

print("Churn Probability (Month-to-month):", probability)
```

**10.2 churn probability for senior citizens**

```python
senior = df[df['SeniorCitizen'] == 1]

# Check if there are any senior citizens to avoid ZeroDivisionError
if len(senior) > 0:
    # 'Churn' column was converted to 0 (No) and 1 (Yes)
    churned_senior = senior[senior['Churn'] == 1]
    probability = len(churned_senior) / len(senior)
    print(f"Churn Probability (Senior Citizens): {probability * 100:.2f}%")
else:
    print("No senior citizens found in the dataset to calculate churn probability.")
```

**10.3 churn probability for costumer without online security**

```python
# Customers without online security: represented by OnlineSecurity_Yes == 0 and OnlineSecurity_No internet service == 0
no_sec = df[(df['OnlineSecurity_Yes'] == 0) & (df['OnlineSecurity_No internet service'] == 0)]

# 'Churn' column was converted to 0 (No) and 1 (Yes)
churned_no_sec = no_sec[no_sec['Churn'] == 1]

# Calculate probability, handle potential ZeroDivisionError
if len(no_sec) > 0:
    probability = len(churned_no_sec) / len(no_sec)
    print(f"Churn Probability (customers without online security): {probability * 100:.2f}%")
else:
    print("No customers found without online security to calculate churn probability.")
```

### **15.3 Retention Strategies**

Based on the analysis, several data-driven retention strategies can be recommended:

1.  **Promote Long-Term Contracts**: Customers on month-to-month contracts have the highest churn rate (42.71%). Offering attractive incentives (discounts, bundled services, loyalty bonuses) for customers to switch to one-year or two-year contracts could significantly reduce churn. Focus on explaining the long-term value and stability of such contracts.

2.  **Optimize Payment Method Experience**: Electronic check users show the highest churn rate (45.29%). Investigate pain points or issues associated with this payment method. Encourage migration to automatic payment methods like bank transfers or credit cards through discounts or added benefits.

3.  **Enhance Service for Fiber Optic Users**: Customers with Fiber optic internet service have a higher correlation with churn. This suggests potential issues with service quality, reliability, or customer support for this segment. Improving the fiber optic experience and proactively addressing potential problems can help retain these customers.

4.  **Boost Online Security Adoption**: Customers without online security features are more prone to churn. Actively promote the benefits of online security, perhaps offering it as a free trial or at a reduced cost, especially to new customers or those identified as high-risk.

5.  **Target Early-Tenure Customers**: New customers (tenure group 'new') are identified as a high-risk group. Implement early engagement programs, welcome packages, or dedicated support to ensure a positive initial experience and build loyalty from the start.

6.  **Review Pricing for High Monthly Charges**: There's an increase in churn rate for customers in the higher monthly charges categories. Conduct a competitive analysis and consider offering personalized plans or discounts to high-spending customers who might be feeling overcharged.

7.  **Personalized Offers for Senior Citizens**: Senior citizens have a higher churn probability. Tailor services, support, and promotions specifically for this demographic, addressing their unique needs and preferences.

8.  **Proactive Churn Prediction**: Utilize predictive models to identify customers likely to churn before they do. This allows for targeted intervention and personalized retention efforts.

### **15.4 Churn Reduction Explanation**

To effectively reduce churn, the telecom company should focus on a multi-faceted approach guided by the insights from this analysis:

1.  **Targeted Retention Campaigns**: Leverage the identification of high-risk customers (e.g., those on month-to-month contracts, using electronic checks, or with Fiber optic internet) to launch specific retention campaigns. These could include personalized offers, proactive customer service outreach, or exclusive benefits to incentivize continued loyalty.

2.  **Product and Service Improvement**: Address the pain points identified with services like Fiber optic internet (if performance or support issues are leading to churn). Enhance online security offerings and clearly communicate their value.

3.  **Customer Lifecycle Management**: Implement strategies for different customer stages:
    *   **Onboarding**: For new customers, provide exceptional onboarding experiences to build early loyalty and reduce early churn.
    *   **Mid-Tenure**: For customers in the 'regular' tenure group, offer loyalty rewards or incentives to upgrade to longer-term contracts.
    *   **Long-Term**: Ensure long-term customers feel valued and continue to receive competitive pricing and service.

4.  **Payment Method Optimization**: Encourage customers to switch from high-churn payment methods (like electronic checks) to more stable, automatic payment options by highlighting convenience or offering small incentives.

5.  **Proactive Engagement**: Utilize predictive models to identify customers who exhibit early signs of churn (e.g., declining usage, increased support calls) and intervene with tailored solutions before they decide to leave.

6.  **Pricing and Value Proposition**: Regularly review pricing strategies, especially for high-spending customers, to ensure perceived value aligns with monthly charges. Consider bundling services that add value and encourage stickiness.

By implementing these data-driven strategies, the telecom company can proactively address the root causes of churn, improve customer satisfaction, and foster long-term customer relationships.
